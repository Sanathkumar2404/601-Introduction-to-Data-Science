```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Inspection = pd.read_excel('39c00379-17c1-4120-96b7-fa372e26f8f0.xlsx')
```

```python
Inspection.head(10)
```

```python
states = pd.read_excel('6481b945-7549-44e5-a65b-8b8eb8b92c87.xlsx')
```

```python
states.head(10)
```

```python
combined_data=pd.merge(Inspection, states, on=["Inspection ID"])
```

```python
combined_data.head(10)
```

```python
data1 =combined_data[combined_data['State']=='Maryland']
```

```python
plt.figure(figsize = (20,10))
sns.catplot(data = data1,x='ProductType',kind='count',aspect = 1)
plt.xticks(rotation=90)
plt.show()
```
