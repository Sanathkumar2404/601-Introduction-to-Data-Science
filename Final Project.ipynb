```python
#Replacing column names as per our convenience
combined_data=combined_data.rename(columns={'Inspection ID':'InspectionId','Legal Name_x':'companyname',
                                            'Product Type':'ProductType'})
```

```python
#Finding the null data in the datasets
combined_data.isnull().sum()
```

```python
#Replacings nulls with zero and unknown
combined_data["Act/CFR Number"].fillna(0, inplace=True)
```

```python
combined_data["Long Description"].fillna('Unknown', inplace=True)
```

```python
combined_data.isnull().sum()
```

<h4>Hypothesis</h4>

1.Maryland has more number of Food Products.


```python
#Retrieving the value count of each state
combined_data['State'].value_counts()
```

```python
#Plotting graph between states versus Product types
plt.figure(figsize = (30,40))
sns.displot(combined_data,x='State',kind='hist',hue='ProductType',aspect = 3)
plt.xticks(rotation=90)
plt.show()
```

```python
#To Focus on our Hypothesis 1 we are looking for only Maryland data
data1 =combined_data[combined_data['State']=='Maryland']
```

```python
#Plotting a graph of maryland Producttype versus count
plt.figure(figsize = (20,10))
sns.catplot(data = data1,x='ProductType',kind='count',aspect = 1)
plt.xticks(rotation=90)
plt.show()
```

By the above graph we can conclude in the mayland the count of Food/cosmetics is more than the other product types therefore the Hypothesis 1 is True.

2.All the companies listed has only one type of product manufacturing unit.

```python
company_df = combined_data[['companyname','ProductType']]
company_df.groupby('companyname').nunique()
```

By the above table we can conclude all the companies are manufacturing only one type of product therefore Hypothesis 2 is True.

1.Which company undergone more number of Inspections?

```python
companycount=combined_data['companyname'].value_counts()
count_df =companycount.to_frame('count')
x=count_df.head(10)
x.plot(kind='bar',color='green')
plt.show
```

By the above graph we can say Titan medical Enterprises, Inc. DBA U.S. Apothecary Labs is the company which undergone more number of Inspections.
